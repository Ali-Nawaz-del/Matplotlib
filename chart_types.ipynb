## Bar chart

```python
# plt.bar(x, height, color = "color name", width = value , label = "label name")
import matplotlib.pyplot as plt

product = ['A','B','C','D']
sales = [1000,1500,800,1200]

plt.bar(product,sales, color = "orange", label = "sales 2025")
plt.xlabel("products")
plt.ylabel("sales")
plt.title("product sales comparison")
plt.legend()

plt.show()
```

## Pie chart

```python
# plt.pie(values , labels = label_list , color = color_list , autopct = '%1.1f%%')
import matplotlib.pyplot as plt

regions = ['north','south','east','west']
revenue = [3000,2000,1500,1000]

plt.pie(revenue, labels = regions, autopct = '%1.1f%%', colors = ['blue','green','red','black'])
plt.title("revenue contribution by region")
plt.show()
```

## Scatter plot (used to find the corelation between two variables)

```python
# plt.scatter(x, y, color = 'color name', label = 'label name', marker = 'marker style')
import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [55,60,65,70,75,80,85,90]

plt.scatter(hours_studies, exam_scores,color = 'green', label = 'student scores', marker = 'o')
plt.xlabel("hours of study")
plt.ylabel("scores of student")
plt.title("Relationship between study time and exam score")
plt.legend()
plt.grid()

plt.show()
```
