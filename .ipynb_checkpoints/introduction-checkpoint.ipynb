```python
import matplotlib.pyplot as plt

x = ['mon','tue','wed','thu']
y = [10,20,5,25]

plt.plot(x,y)
plt.title("sales this week")
plt.xlabel("days of weed")
plt.ylabel("sales per day")
plt.show()
```
