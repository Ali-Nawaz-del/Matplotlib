```python
# plt.subplot(nrows, ncols, index)

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,30,20,15]

plt.subplot(1,2,1) # 1 row, 2 colunm, 1st subplot
plt.plot(x,y)
plt.title("line chart")

plt.subplot(1,2,2) # 1 row, 2 colunm, 2nd subplot
plt.bar(x,y)
plt.title("bar chart")

plt.tight_layout()
plt.show()
```

```python
# fig, ax = plt.subplots(nrows, ncols, figsize = (width, hight))

import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2, figsize = (10,5))

x = [1,2,3,4]
y = [10,30,20,15]

ax[0].plot(x,y, color = "green")
ax[0].set_title("line plot")

ax[1].bar(x,y, color = "blue")
ax[1].set_title("bar chart")

fig.suptitle("comparison of line and bar chart")

plt.tight_layout()
plt.show()
```
